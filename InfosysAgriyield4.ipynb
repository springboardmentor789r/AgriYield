```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/crop_yield_dataset.csv")

plt.figure(figsize=(8,5))
plt.hist(df['Temperature'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Temperature")
plt.xlabel("Temperature(C)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(x='Humidity', y='Crop_Yield', data=df)
plt.title("Humidity vs Crop Yield")
plt.show()
```

```python
sns.lmplot(x='Humidity', y='Crop_Yield', data=df, line_kws={"color": "red"})
plt.title("Humidity vs Crop Yield (with Trend Line)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for plots
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))

# Plot histogram for every numerical column
df.hist(figsize=(15, 10), bins=20, edgecolor='black')
plt.suptitle(" Distribution of All Numerical Fields", fontsize=16)
plt.show()
```

```python
# Pairplot - compare all numerical fields against Crop_Yield
sns.pairplot(df, vars=['Temperature', 'Humidity', 'N','P','K'], diag_kind='hist')
plt.suptitle("Pairwise Relationships between Features and Crop Yield", y=1.02)
plt.show()

# Individual scatter plots for better clarity
features = ['Temperature', 'Humidity', 'N','P','K']
for col in features:
    plt.figure(figsize=(7, 4))
    sns.scatterplot(x=df[col], y=df['Crop_Yield'])
    plt.title(f"{col} vs Crop Yield")
    plt.xlabel(col)
    plt.ylabel("Crop Yield")
    plt.show()
```

```python
if 'Soil_Type' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Soil_Type', y='Crop_Yield', data=df)
    plt.title("Crop Yield by Soil Type")
    plt.xlabel("Soil Type")
    plt.ylabel("Crop Yield")
    plt.show()

    sns.countplot(x='Soil_Type', data=df)
    plt.title("Distribution of Soil Types")
    plt.show()

if 'Crop_Type' in df.columns:
    plt.figure(figsize=(10, 7))
    sns.boxplot(x='Crop_Type', y='Crop_Yield', data=df)
    plt.title("Crop Yield by Crop Type")
    plt.xlabel("Crop Type")
    plt.ylabel("Crop Yield")
    plt.show()

    # Count plot
    sns.countplot(x='Crop_Type', data=df)
    plt.title("Distribution of Crop Types")
    plt.show()

```

```python
# Example: Soil_Type pie chart
if 'Soil_Type' in df.columns:
    plt.figure(figsize=(6, 6))
    df['Soil_Type'].value_counts().plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        colors=sns.color_palette('pastel')
    )
    plt.title("Distribution of Soil Types", fontsize=14)
    plt.ylabel("")  # Hide y-label
    plt.show()

# Example: Crop_Type pie chart (if exists)
if 'Crop_Type' in df.columns:
    plt.figure(figsize=(6, 6))
    df['Crop_Type'].value_counts().plot.pie(
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        colors=sns.color_palette('cool')
    )
    plt.title("Distribution of Crop Types", fontsize=14)
    plt.ylabel("")
    plt.show()
```
