```python
import matplotlib.pyplot as plt

# Histogram
plt.hist(df["Crop_Yield"], bins=30)
plt.title("Crop Yield Distribution")
plt.xlabel("Crop Yield")
plt.ylabel("Count")
plt.show()
```

```python
# Bar Graph
avg_yield = df.groupby("Crop_Type")["Crop_Yield"].mean()
plt.bar(avg_yield.index, avg_yield.values)
plt.title("Average Crop Yield by Crop Type")
plt.xlabel("Crop Type")
plt.ylabel("Average Yield")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot
plt.scatter(df["Temperature"], df["Crop_Yield"])
plt.title("Temperature vs Crop Yield")
plt.xlabel("Temperature")
plt.ylabel("Crop Yield")
plt.show()
```

```python
# Line plot
plt.plot(df["Date"], df["Temperature"])
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.show()
```

```python
# Box plot
soil_types = df["Soil_Type"].unique()
data = [df[df["Soil_Type"] == soil]["Soil_pH"] for soil in soil_types]

plt.boxplot(data, tick_labels=soil_types)
plt.title("Soil pH by Soil Type")
plt.xlabel("Soil Type")
plt.ylabel("Soil pH")
plt.show()
```
