## Encoding

```python
# Label Encoded
df2=df.copy(deep=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le_crop=LabelEncoder()
le_soil=LabelEncoder()
df2['crop_type_encoded']=le_crop.fit_transform(df2['crop_type'])
df2['soil_type_encoded']=le_soil.fit_transform(df2['soil_type'])
df2
```

```python
# OneHotEncoded
from sklearn.preprocessing import OneHotEncoder
df3=df.copy(deep=True)
encoder = OneHotEncoder(sparse_output=False)
categorical_cols=['crop_type','soil_type']
encoded = encoder.fit_transform(df3[categorical_cols])
encoded_columns = encoder.get_feature_names_out(categorical_cols)
encoded_df = pd.DataFrame(encoded, columns=encoded_columns, index=df3.index)
df_encoded = pd.concat([df3.drop(columns=categorical_cols), encoded_df], axis=1)
df3.drop(columns=categorical_cols, inplace=True)
df3 = pd.concat([df3, encoded_df], axis=1)
df3
```

```python
# Binary Encoders
import category_encoders as ce
df4= df.copy(deep=True)
encoder = ce.BinaryEncoder(cols=categorical_cols)
encoded_cols = encoder.fit_transform(df4[categorical_cols])
df4.drop(columns=categorical_cols, inplace=True)
df4 = pd.concat([df4, encoded_cols], axis=1)
df4
```

```python
# Target Encoding

target_col = 'crop_yield'  
categorical_cols = ['crop_type', 'soil_type']
df5 = df.copy(deep=True)
encoder = ce.TargetEncoder(cols=categorical_cols)
df5[categorical_cols] = encoder.fit_transform(df5[categorical_cols], df5[target_col])
df5
```

```python
# Leave One Out Encoding

target_col = 'crop_yield'
categorical_cols = ['crop_type', 'soil_type']
df6=df.copy(deep=True)
encoder = ce.LeaveOneOutEncoder(cols=categorical_cols)
df6[categorical_cols] = encoder.fit_transform(df6[categorical_cols], df6[target_col])
df6
```

## Scaling

```python
# Extract year, month, day from date

df3['date'] = pd.to_datetime(df3['date'])
df3['year'] = df3['date'].dt.year
df3['month'] = df3['date'].dt.month
df3['day'] = df3['date'].dt.day
# Drop original date column
df3.drop('date', axis=1, inplace=True)
df3
```

```python
# Create copies for scaling
df_scale1=df3.copy(deep=True)
df_scale2=df3.copy(deep=True)
```

```python
df_scale1.columns
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
# Define columns to be scaled
columns_to_scale=['soil_ph','temp','humidity','wind_speed','n','p','k','soil_quality']
```

```python
columns_to_scale
```

```python
# ----- Min-Max Scaling -----
scaler=MinMaxScaler()
df_scale1[columns_to_scale]=scaler.fit_transform(df_scale1[columns_to_scale])
df_scale1
```

```python
# ----- Standard Scaling -----
scaler=StandardScaler()
df_scale2[columns_to_scale]=scaler.fit_transform(df_scale2[columns_to_scale])
df_scale2
```

```python
df_scale1.to_csv('Crop_Mod_OneH_MinMax.csv',index=False)
```

```python
df_scale2.to_csv('Crop_Mod_OneH_Stand.csv',index=False)
```
